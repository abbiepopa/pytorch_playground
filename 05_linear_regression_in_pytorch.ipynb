```python
import torch
from torch.autograd import Variable
```

```python
x_data = Variable(torch.Tensor([[1.0], [2.0], [3.0]]))
y_data = Variable(torch.Tensor([[2.0], [4.0], [6.0]]))
```

```python
class Model(torch.nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.linear = torch.nn.Linear(1,1)
        
    def forward(self, x):
        y_pred = self.linear(x)
        return y_pred
    
model = Model()
```

```python
model
```

```python
criterion = torch.nn.MSELoss(reduction='sum') 
#size_average is for averaging across the batch, we used to set size_average=Flase in this case, but that
#will be deprecated so now we use reduction='sum'
optimizer = torch.optim.SGD(model.parameters(), lr = 0.01)
```

```python
criterion
```

```python
optimizer
```

```python
hour_var = Variable(torch.Tensor([[4.0]]))
print('predict (before training)', 4, model.forward(hour_var).data[0][0])
```

```python
for epoch in range(500):
    y_pred = model(x_data)
    loss = criterion(y_pred, y_data)
    print(epoch, loss.data)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
print('predict (after training)', 4, model.forward(hour_var).data[0][0])
```
